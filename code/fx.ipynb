```python
def read_fr_file(file_name):
    cls = []
    a = []
    with open(file_name, 'r') as f:
        for line in f.readlines():
            l = line.strip().split(' ')
            cls.append(' '.join(l[:-1]))
            a.append(float(l[-1]))
    return cls, a
```

```python
# %%
import json
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt


def plot_curve_scatter(data, names, save_name=None, metric='avg'):
    # for k, v in data.items():
    #     print(k, len(v))
    # df = pd.DataFrame(data)

    # Plotting
    plt.rcParams.update({'font.size': 17})
    
    fig, ax = plt.subplots()
    colors = ['#1F77B4', '#FF7F0E','green']  # Deep blue and bright yellow
    # colors = ['#F8A12D', '#2DA4CD','#A9D18E']  # Deep blue and bright yellow
    markers = ['o', '.','2']
    print(names)
    markers = markers[:len(names)]
    # Scatter plot for each model type
    for i, (model, marker, color) in enumerate(zip(names, markers, colors)):
        # print(i)
        ax.scatter(data[i][0], data[i][1],marker=marker, label=model, s=100, color=color)  # Marker size = 100
        # ax.plot(subset['avg_bit'], subset['accuracy'], linestyle='--', linewidth=3, color=color)  # Dashed line
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x:.1f}"))
    t = 'AP50'
    if metric == 'avg':
        ax.set_ylabel(t,fontsize=20)
    else:
        ax.set_ylabel(t,fontsize=20)
    ax.set_xlabel(t,fontsize=20)
    # ax.legend()
    ax.grid(True)
    # plt.legend(loc='upper left')
    # plt.legend(loc='upper center',ncol=4,fontsize=15)
    
    plt.tight_layout()
    if save_name is None:
        plt.show()
    else:
        plt.savefig(save_name,transparent=True)
```
