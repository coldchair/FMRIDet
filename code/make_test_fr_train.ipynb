```python
import numpy as np
dataset_type = 'CocoNSDDataset'
data_root = './'
image_dir = '../../nsd_processed_data/all_images'

# default : each sample

subj = 'subj01'
SAVE_ROOT_DIR = '../../nsd_processed_data'
ann_file = f'{SAVE_ROOT_DIR}/instances_0_73000_0.01.json'
index_file_tr = f'{SAVE_ROOT_DIR}/mrifeat/{subj}/index_each_tr.npy'
index_file_te = f'{SAVE_ROOT_DIR}/mrifeat/{subj}/index_each_te.npy'

# early ventral midventral midlateral lateral parietal
rois = ['early', 'ventral', 'midventral', 'midlateral', 'lateral', 'parietal']

fmri_files_path_tr = []
fmri_files_path_te = []

for roi in rois:
    fmri_files_path_tr.append(f'{SAVE_ROOT_DIR}/mrifeat/{subj}/{subj}_{roi}_betas_tr.npy')
    fmri_files_path_te.append(f'{SAVE_ROOT_DIR}/mrifeat/{subj}/{subj}_{roi}_betas_te.npy')

X = []
for fmri_file in fmri_files_path_tr:
    cX = np.load(fmri_file).astype("float32")
    X.append(cX)
X = np.hstack(X) # shape : (n_samples, n_voxels)

img_ids = np.load(index_file_tr).tolist()

print(img_ids)
```

```python
img_ids = img_ids[:50]
for roi in rois:
    output_name = f'{SAVE_ROOT_DIR}/mrifeat/{subj}/{subj}_{roi}_betas_tr_50.npy'
    input_name = f'{SAVE_ROOT_DIR}/mrifeat/{subj}/{subj}_{roi}_betas_tr.npy'
    cX = np.load(input_name).astype("float32")
    cX = cX[:50]
    print(cX.shape)
    np.save(output_name, cX)
index_file_tr_50 = f'{SAVE_ROOT_DIR}/mrifeat/{subj}/index_each_tr_50.npy'
np.save(index_file_tr_50, img_ids)
```
